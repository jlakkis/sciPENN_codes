```python
import pandas as pd
from os.path import join
```

```python
base = 'runtime'

totalvi = pd.read_csv(join(base, "totalVI.csv"), index_col = 0)
scipenn = pd.read_csv(join(base, "scipenn.csv"), index_col = 0)
seurat = pd.read_csv(join(base, "seurat.csv"), index_col = 0)
```

```python
runtime = pd.concat((totalvi, scipenn, seurat), axis = 1).plot.line()
runtime.set_ylabel("runtime (seconds)")
runtime.set_xlabel("fraction of data")
```

```python
runtime.figure.savefig('runtimefig.pdf')
```
